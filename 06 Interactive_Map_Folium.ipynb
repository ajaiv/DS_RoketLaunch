Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
# site_map.add_child(marker_cluster) - as in original code
# Create a MarkerCluster object and add it to the map
marker_cluster = MarkerCluster().add_to(site_map)

# Add markers for each launch result in spacex_df to the MarkerCluster
for index, record in spacex_df.iterrows():
    coordinate = [record['Lat'], record['Long']]
    
    # Create a custom icon with the color based on the marker_color column
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(
            color='white',  # Marker outline color
            icon_color=record['marker_color'],  # Marker icon color (green/red)
            icon='info-sign'  # You can change the icon type as needed
        ),
        popup=folium.Popup(f"Launch Site: {record['Launch Site']}<br>Class: {'Success' if record['class'] == 1 else 'Failure'}")
    )
    
    # Add the marker to the marker_cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians
import folium

# Function to calculate the distance between two geographical points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth radius in km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordinates of the launch site (example: KSC LC-39A)
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# Create the map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add MousePosition to get coordinates
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Mark the launch site with a green marker
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site: KSC LC-39A",
    icon=folium.Icon(color='green')
).add_to(site_map)

# Define coordinates of a city, railway, highway, or other points of interest (example: nearest city)
city_lat = 28.5383  # Example city latitude (Orlando, FL)
city_lon = -81.3792  # Example city longitude

# Calculate the distance from the launch site to the city
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
print(f"Distance from launch site to city: {distance_city:.2f} km")

# Mark the city with a red marker
folium.Marker(
    location=[city_lat, city_lon],
    popup=f"City: Orlando<br>Distance: {distance_city:.2f} km",
    icon=folium.Icon(color='red')
).add_to(site_map)

# Draw a line (PolyLine) between the launch site and the city
coordinates = [
    [launch_site_lat, launch_site_lon],  # Launch site coordinates
    [city_lat, city_lon]                 # City coordinates
]

# Create PolyLine object
lines = folium.PolyLine(locations=coordinates, weight=2, color='blue').add_to(site_map)

# Display the map
site_map
```
